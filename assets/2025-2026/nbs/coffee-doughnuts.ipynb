```python
import numpy as np
import matplotlib.pyplot as plt

# Define ranges for x1 and x2
x1 = np.linspace(-12, 30, 400)
x2 = np.linspace(-8, 10, 400)

# Create a meshgrid
X1, X2 = np.meshgrid(x1, x2)

# Compute the function
F = np.log(1 + X1) + 2 * np.sqrt(X2) * (13 - X1) + 2 * np.sqrt(8 - X2)

# Plot the surface
plt.figure(figsize=(8, 6))
contour = plt.contourf(X1, X2, F, levels=50, cmap="viridis")
plt.colorbar(contour, label=r"$f(x_1, x_2)$")
plt.xlabel(r"$x_1$")
plt.title(r"$f(x_1, x_2) = \log(1 + x_1) + 2\sqrt{x_2}(13 - x_1) + 2\sqrt{8 - x_2}$")
plt.axhline(8, color="red")
plt.axhline(0, color="red")
plt.axvline(12, color="red")
plt.axvline(0, color="red")
plt.scatter(1.243,6.869)
plt.show()
```
