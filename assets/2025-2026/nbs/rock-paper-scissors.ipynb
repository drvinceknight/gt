```python
import numpy as np
import nashpy as nash
import matplotlib.pyplot as plt

M_r = np.array([[0, 1, -1], [-1, 0, 1], [1, -1, 0]])
rock_paper_scissors = nash.Game(M_r)
```

```python
rock_paper_scissors
```

```python
x0 = np.array([1 / 6, 1 / 6, 2 / 3])
timepoints = np.linspace(0, 10, 3000)
xs = rock_paper_scissors.replicator_dynamics(y0=x0, timepoints=timepoints).T
xs
```

```python
plt.figure()
plt.plot(timepoints, xs.T)
plt.ylim(0, 1)
plt.legend(["$x_1$", "$x_2$", "$x_3$"])
plt.ylabel("Distribution")
plt.xlabel("Time");
```
