```python
import numpy as np
import nashpy as nash
```

```python
A = np.array(((2, -2), (-1, 1)))
B = - A
game = nash.Game(A, B)
```

```python
game
```

```python
sigma_r = np.array((.2, .8))
sigma_c = np.array((1, 0))
```

```python
game.is_best_response(sigma_r=sigma_r, sigma_c=sigma_c)
```

```python
sigma_r = np.array((1 / 3, 2 / 3))
sigma_c = np.array((1 / 2, 1 / 2))
game.is_best_response(sigma_r=sigma_r, sigma_c=sigma_c)
```
